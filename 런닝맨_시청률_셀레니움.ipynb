```python
viewr_ratings = [] # 시청률 넣을 리스트
episode_list = [] # 회차 넣을 리스트
```

위키백과에서 런닝맨 시청률은 총 4가지가 나온다

4가지 중 AGB 시청률이 닐슨 코리아에서 제공한 값이므로

이 값만 가져온다

```python
# 2021년도 시청률 가져옴 - 4가지
xpath_basic = """//*[@id="mw-content-text"]/div[1]/div[14]/div[2]/table/tbody/tr[""" # xpath 기본

episode_count = 1
for i in tqdm(np.arange(2, 46)):
    # 회차
    if i <= 27:
        tmp1 = i + episode_count
        episode_count += 1
    else:
        tmp1 = i + 27
    episode_tmp = str(tmp1) + "]/th" # 각 회차에 얻어오가 위함
    episode_xpath = xpath_basic + episode_tmp # 회차 얻어오기 위한 xpath값 구함
    
    find_element = driver.find_element_by_xpath(episode_xpath)
    episode = find_element.get_attribute('innerHTML')
    episode = episode.replace("\n", "")
    episode_list.append(episode)
    
    # 시청률
    if i <= 27:
        tmp2 = i * 2
        viewr_tmp = str(tmp2) + "]/td[2]" # 각 시청률에 얻어오가 위함
    else:
        tmp2 = i + 27
        viewr_tmp = str(tmp2) + "]/td[3]" # 각 시청률에 얻어오가 위함
    viewr_xpath = xpath_basic + viewr_tmp # 시청률 얻어오기 위한 xpath값 구함
    
    find_element = driver.find_element_by_xpath(viewr_xpath)
    viewr = find_element.get_attribute('innerHTML')
    viewr = viewr.replace("\n", "")
    viewr_ratings.append(viewr)
```

```python
# 2021년도 시청률 가져옴 - 4가지
xpath_basic = """//*[@id="mw-content-text"]/div[1]/div[13]/div[2]/table/tbody/tr[""" # xpath 기본

episode_count = 1
for i in tqdm(np.arange(0, 10)):
    # 회차
    tmp1 = 87 + (i * 2)
    episode_tmp = str(tmp1) + "]/th" # 각 회차에 얻어오가 위함
    episode_xpath = xpath_basic + episode_tmp # 회차 얻어오기 위한 xpath값 구함
    
    find_element = driver.find_element_by_xpath(episode_xpath)
    episode = find_element.get_attribute('innerHTML')
    episode = episode.replace("\n", "")
    episode_list.append(episode)
    
    # 시청률
    tmp2 = 88 + (i * 2)
    viewr_tmp = str(tmp2) + "]/td[2]" # 각 시청률에 얻어오가 위함
    viewr_xpath = xpath_basic + viewr_tmp # 시청률 얻어오기 위한 xpath값 구함
    
    find_element = driver.find_element_by_xpath(viewr_xpath)
    viewr = find_element.get_attribute('innerHTML')
    viewr = viewr.replace("\n", "")
    viewr_ratings.append(viewr)
```

```python
# 회차, 시청률 저장할 데이터프레임
runningman_viewr = pd.DataFrame(columns=['회차', '시청률', '회차_merge용'])
runningman_viewr
```

```python
# 데이터프레임에 저장
for i in np.arange(0, len(viewr_ratings)):
    merge_episode = "런닝맨 " + str(episode_list[i]) + "회"
    dic = {"회차" : episode_list[i],
           "시청률" : viewr_ratings[i],
           "회차_merge용" : merge_episode}
    
    runningman_viewr = runningman_viewr.append(dic, ignore_index=True)

# 시청률 순으로 내림차순 정렬
runningman_viewr = runningman_viewr.sort_values('회차', ascending=False).reset_index(drop=True)
runningman_viewr
```

```python
# 시청률 csv 파일로 저장
runningman_viewr.to_csv('runningman_viewr.csv', index=False, encoding="utf-8-sig")
```
